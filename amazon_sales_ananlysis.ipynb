```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load the dataset
data = pd.read_csv('data/Amazon Sale Report.csv')
```

```python

# Function to normalize state names
def normalize_state_names(state_series):
    state_series = state_series.str.upper()  # Convert to uppercase for consistency
    state_series = state_series.str.strip()  # Remove leading/trailing whitespace
    return state_series

# Normalize state names
data['ship-state'] = normalize_state_names(data['ship-state'])
```

```python
# Ensure the 'visualizations' directory exists
if not os.path.exists('visualizations'):
    os.makedirs('visualizations')
```

```python
# Sales Overview
def sales_overview(data):
    sales_trends = data.groupby('Date')['Amount'].sum()
    plt.figure(figsize=(25, 6))
    plt.plot(sales_trends.index, sales_trends.values, marker='o')
    plt.title('Sales Trends Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Sales Amount')
    plt.xticks(rotation=45)
    plt.savefig('visualizations/sales_overview.png')
    plt.show()
```

```python
# Product Analysis
def product_analysis(data):
    product_dist = data['Category'].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=product_dist.index, y=product_dist.values, palette='viridis')
    plt.title('Product Category Distribution')
    plt.xlabel('Product Category')
    plt.ylabel('Quantity Sold')
    plt.xticks(rotation=45)
    plt.savefig('visualizations/product_analysis.png')
    plt.show()
```

```python
# Fulfillment Analysis
def fulfillment_analysis(data):
    fulfillment_dist = data['Fulfilment'].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=fulfillment_dist.index, y=fulfillment_dist.values, palette='muted')
    plt.title('Fulfillment Methods Distribution')
    plt.xlabel('Fulfillment Method')
    plt.ylabel('Number of Orders')
    plt.xticks(rotation=45)
    plt.savefig('visualizations/fulfillment_analysis.png')
    plt.show()
```

```python
# Customer Segmentation
def customer_segmentation(data):
    customer_dist = data['ship-state'].value_counts()
    plt.figure(figsize=(20, 6))
    sns.barplot(x=customer_dist.index, y=customer_dist.values, palette='coolwarm')
    plt.title('Customer Segmentation by State')
    plt.xlabel('State')
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45)
    plt.savefig('visualizations/customer_segmentation.png')
    plt.show()
```

```python
# Geographical Analysis
def geographical_analysis(data):
    sales_by_state = data.groupby('ship-state')['Amount'].sum()
    plt.figure(figsize=(20, 6))
    sns.barplot(x=sales_by_state.index, y=sales_by_state.values, palette='crest')
    plt.title('Sales Distribution by State')
    plt.xlabel('State')
    plt.ylabel('Total Sales Amount')
    plt.xticks(rotation=45)
    plt.savefig('visualizations/geographical_analysis.png')
    plt.show()
```

```python
# Running the analysis functions
sales_overview(data)
product_analysis(data)
fulfillment_analysis(data)
customer_segmentation(data)
geographical_analysis(data)
```
